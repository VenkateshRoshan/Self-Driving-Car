* `Motor` - `RPI 4` - `L293D` Connections

---
* Connect L293D 8 and 16 pins with 5V with RPI 4 pin #(2/4)
* Connect L293D GND pin (4/5/12/13) with GND pin of RPI 4 pin #(6/9/14/20/25/30/34/39)
* Connect L293D Input1 pin #2 with RPI 4 GPIO 2 (or) pin #3 (In program we have to make Pin3 3 is the GPIO.setup(3,GPIO.OUT))
* Connect L293D Output1 Pin #3 with Motor +ve `Forward Direction`
* Connect L293D Input2 pin #7 with RPI 4 GPIO 3 (or) pin #5
* Connect L293D Output2 Pin #6 with Motor -ve `Backward Direction`
* Connect L293D Enable Pin #1 with RPI 4 GPIO 4 pin #7
